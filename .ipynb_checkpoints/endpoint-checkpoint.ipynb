```python
import yaml
data = """
version: v.0.0.1
endpoint:
  dt_company:
    insert:
      fields:
        id_company:
          name: primary key database
          type: int
        nm_company:
          name: company name
          type: string
        url_company:
          name: company url
          type: string
    where:
      tags:
        id_company:
          name: primary key database
          type: int
        nm_company:
          name: company name
          type: string
    remove:
      tags:
        id_company:
          name: primary key database
          type: int
    update:
      tags:
        id_company:
          name: primary key database
          type: int
      fields:
        nm_company:
          name: company name
          type: string
        url_company:
          name: company url
          type: string        
  dt_product:
    insert:
      fields:
        id_product:
          name: primary key
          type: int
        nm_product:
          name: product name
          type: str
        nm_databaseDetail:
          name: Database reference
          type: str
    where:
      tags:
        id_product:
          name: primary key
          type: int
        nm_product:
          name: product name
          type: str
        nm_databaseDetail:
          name: Database reference
          type: str
    remove:
      tags:
        id_product:
          name: primary key
          type: int
    update:
      tags:
        id_product:
          name: primary key
          type: int
      fields:
        id_product:
          name: primary key
          type: int
        nm_product:
          name: product name
          type: str
        nm_databaseDetail:
          name: Database reference
          type: str
  dt_company_product:
    insert:
      fields:
        id_company_product:
          name: primary key
          type: int
        id_product:
          name: foreign key
          type: int
        id_company:
          name: foreign key
          type: int
        nm_product:
          name: product name
          type: str
        id_worker:
          name: worker reference
          type: int
    where:
      tags:
        id_company_product:
          name: primary key
          type: int
        id_product:
          name: foreign key
          type: int
        id_company:
          name: foreign key
          type: int
        nm_product:
          name: product name
          type: str
        id_worker:
          name: worker reference
          type: int
    remove:
      tags:
        id_company_product:
          name: primary key
          type: int
    update:
      tags:
        id_company_product:
          name: primary key
          type: int
      fields:
        id_company_product:
          name: primary key
          type: int
        id_product:
          name: foreign key
          type: int
        id_company:
          name: foreign key
          type: int
        nm_product:
          name: product name
          type: str
        id_worker:
          name: worker reference
          type: int
  dt_vm:
    insert:
      fields:
        id_vm:
          name: primary_key
          type: int
        vCPU:
          name: vCPU in Core
          type: str
        clock:
          name: Clock in GHz
          type: str
        RAM:
          name: RAM
          type: str
        OS:
          name: OS_TYPE
          type: str
        Storage_Volume:
          name: Storage Data in GB
          type: str
        SSD_Volume:
          name: SSD Storage in GB
          type: str
        Snapshot_Volume:
          name: Snapshot Storage in GB
          type: str
        Template_Volume:
          name: Template Storage in GB
          type: str
        ISO_Volume:
          name: ISO Volume in GB
          type: str
        Public IP:
          name: Public IP Available
          type: str
        Backup_Storage:
          name: Backup Storage Available
          type: str
        Price :
          name: Price in Rp/Month
          type: str
        id_company_product:
          name: fk_dt_company_product
          type: int
        id_company:
          name: fk_dt_company
          type: int
    where:
      tags:
        id_vm:
          name: primary_key
          type: int
        vCPU:
          name: vCPU in Core
          type: str
        clock:
          name: Clock in GHz
          type: str
        RAM:
          name: RAM
          type: str
        OS:
          name: OS_TYPE
          type: str
        Storage_Volume:
          name: Storage Data in GB
          type: str
        SSD_Volume:
          name: SSD Storage in GB
          type: str
        Snapshot_Volume:
          name: Snapshot Storage in GB
          type: str
        Template_Volume:
          name: Template Storage in GB
          type: str
        ISO_Volume:
          name: ISO Volume in GB
          type: str
        Public IP:
          name: Public IP Available
          type: str
        Backup_Storage:
          name: Backup Storage Available
          type: str
        Price :
          name: Price in Rp/Month
          type: str
        id_company_product:
          name: fk_dt_company_product
          type: int
        id_company:
          name: fk_dt_company
          type: int
    remove:
      id_vm:
        name: primary_key
        type: int
    update:
        id_vm:
          name: primary_key
          type: int
        vCPU:
          name: vCPU in Core
          type: str
        clock:
          name: Clock in GHz
          type: str
        RAM:
          name: RAM
          type: str
        OS:
          name: OS_TYPE
          type: str
        Storage_Volume:
          name: Storage Data in GB
          type: str
        SSD_Volume:
          name: SSD Storage in GB
          type: str
        Snapshot_Volume:
          name: Snapshot Storage in GB
          type: str
        Template_Volume:
          name: Template Storage in GB
          type: str
        ISO_Volume:
          name: ISO Volume in GB
          type: str
        Public IP:
          name: Public IP Available
          type: str
        Backup_Storage:
          name: Backup Storage Available
          type: str
        Price :
          name: Price in Rp/Month
          type: str
        id_company_product:
          name: fk_dt_company_product
          type: int
        id_company:
          name: fk_dt_company
          type: int
"""


def repodata():
    d_model = yaml.safe_load(data)
    return d_model

def parser(json, command=None):
    sdl_endpoint = repodata()
    endpoint = sdl_endpoint['endpoint'][command]
    parameter_fix = list()
    key_query = None
    return_paramfix = list()
    for i in json:
        tagfields=dict()
        for parameters_key in endpoint[i]:
            try:
                parameters_check = json[i][parameters_key]
            except Exception:
                parameters_check = None
            if parameters_check:
                key_query = parameters_key
                data_variabel = dict()
                paramemeters_data = endpoint[i][parameters_key]
                for variabel_key in paramemeters_data: 
                    try:
                        variabel_check = json[i][parameters_key][variabel_key]
                    except Exception:
                        variabel_check = None
                    if variabel_check:
                        data_variabel[variabel_key] = json[i][parameters_key][variabel_key]
                    else:
                        data_variabel[variabel_key] = endpoint[i][parameters_key][variabel_key]['default']
                    tagfields[parameters_key]=data_variabel
        parameter=dict()
        for key in tagfields:
            try:
                fields_check = tagfields["fields"]
            except Exception:
                fields_check = None

            try:
                tags_check = tagfields["tags"]
            except Exception:
                tags_check = None

            if fields_check is None and tags_check is not None:
                parameter = {
                    'table': command,
                    'tags' : tagfields['tags']
                }
            elif fields_check is not None and tags_check is None:
                parameter = {
                    'table': command,
                    'fields' : tagfields['fields']
                }
            elif fields_check is not None and tags_check is not None:
                parameter = {
                    'table': command,
                    'fields' : tagfields['fields'],
                    'tags' : tagfields['tags']
                }
            else:
                data = {
                    key_query : tagfields[key]
                }
                parameter = {
                    'table': command,
                    "query" : data,
                }

        parameter_fix.append(parameter)
        return_paramfix = {
            'action': i,
            'data': parameter_fix,
        }
    return return_paramfix

```
